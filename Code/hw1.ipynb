```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Question 2.a

T = 1/10 # seconds
time = np.linspace(0,2*T,200)
displacement = [y for y in np.cos(((2*np.pi)/T)*time)]
plt.scatter(time, displacement);
```

```python
# Question 3

f = 1 # Frequency Hz
w = 2*np.pi*f # Angualar Frequency rad/s

time = np.linspace(0,2,200) # From time 0 to 2*T

y1 = [y for y in 3*np.cos(w*time)]
y2 = [y for y in 3*np.sin(w*time+(np.pi/3))]

plt.scatter(time, y1, label='y1')
plt.scatter(time, y2, label='y2')
plt.legend()
plt.show;
```
